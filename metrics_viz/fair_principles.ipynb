## F
A global, unique and persistent identifier is achieved, amon others, by either having a GitHub release or a published Pypi package (pip installable).

```python
input_file = "../data/pip_repo_combo.csv"
metadata = pd.read_csv(input_file, low_memory=False)
metadata.set_index('Unnamed: 0', inplace=True)

release_count = np.load("../data/release_count.npy")

val_pip, count_pip = np.unique(metadata["exists_in_pypi"], return_counts=True)
val_release, count_release = np.unique(np.where(release_count>0, True, False), return_counts=True)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].pie(count_pip, labels=val_pip)
ax[1].pie(count_release, labels=val_release)
```
