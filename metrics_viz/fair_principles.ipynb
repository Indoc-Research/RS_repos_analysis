A global, unique and persistent identifier is achieved, amon others, by either having a GitHub release or a published Pypi package (pip installable). To estimate the percentage of repos with a pip release, we only considered repos for which the first language was Python.

```python
input_file = "../data/pip_repo_combo.csv"
pip_dataset = pd.read_csv(input_file, low_memory=False)
pip_dataset.set_index('Unnamed: 0', inplace=True)
pip_dataset = pip_dataset[pip_dataset["language"]=="Python"]
release_count = np.load("../data/release_count.npy")

val_pip, count_pip = np.unique(pip_dataset["exists_in_pypi"], return_counts=True)
val_release, count_release = np.unique(np.where(release_count>0, True, False), return_counts=True)
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].pie(count_pip, labels=val_pip, autopct='%1i%%')
ax[1].pie(count_release, labels=val_release,  autopct='%1i%%')
ax[0].set_title("Proportion of repos on PyPI")
ax[1].set_title("Proportion of repos with releases")
plt.savefig("../figures/pip_v_releases.png")
```
